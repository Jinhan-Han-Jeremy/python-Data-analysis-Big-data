# Algorithm A: Iterative Approach in Linear Time

```python
import pandas as pd
count = 0           # total number of operations executed in f(n)
def f(n):
    """f(n) = f(n-1)+f(n-2)"""
    global count
    x = 0           # correction from 1 to 0
    y = 1
    z = n
    count = 0
    #Given x, y, z
    
    for i in range (2, n+1):  # xxx correction from n to n+1
        z = x+y
        x = y
        y = z
        count += 1
        
    return z

n=30

"""
"""

```

# Approach B: Recursive Approach in Exponential Time

```python
import pandas as pd
count = 0
def g(n):
    """g(n) = g(n-1)+g(n-2)"""
    global count
    if n <2:
        return n
    count += 1
    return g(n-1)+g(n-2)

n=30



d = [ [1,2,3], [4,5,6], [7,8,9]]
df = pd.DataFrame(d)

w = df.loc [2 ]
x = df.loc[ [2] ]
y = df [ 2 ] 
z = df [ [2] ] 

```

# Approach C: Recursive Matrix Approach in Logarithmic Time

* See details at https://mathworld.wolfram.com/FibonacciQ-Matrix.html

```python
import pandas as pd
d = [ [1,1], [1,0] ]
q = pd.DataFrame (d)

count = 0
def h(q,n):
    global count
    if n <=1:
        return q
    r = h(q,n//2)             # r= (q**n//2)
    r = r@r; count += 1       # q**n = r*r 
    if n%2==1:
        r = r@q; count += 1   # q**n = r*r*q. if n%2==1
    return r

n=30
```

# (5pt) Q3: Display the curve showing the run-time increases most slowly as a logarithmic funciton of n as n increases from 0 to 30.
## Below is the expected output.

```python
from IPython.display import Image
Image(filename='algo.C.png')
```

```python
d = {"index":[], "count_C":[],"fib_C":[]}

for i in range(n):
        count = 0
        x= h(q,i)
        d["index"].append(i)
        d["fib_C"].append(x[0][1])
        d["count_C"].append(count)


df3 = pd.DataFrame(d)
d = df3["count_C"].plot(title = "Algorithm C: Logarithmic Time Coplexity",legend = True)
#count_C using h(q,i) shows graph
df3
```
